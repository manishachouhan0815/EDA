```python
df=pd.DataFrame(np.random.randn(100,3), columns=['A','B','C'])
df.head()
df['A']=df['A'].cumsum()+20
df['B']=df['B'].cumsum()+20
df['C']=df['C'].cumsum()+20
```

```python
df.head()
```

```python
df.iplot()
#it is a scatter lot in plotly
```

```python
plt.plot(df)
#better version of above
```

```python
df.iplot(x='A', y='B', mode='markers', size=25)
```

```python
df.iplot(x='A', y='B', mode='lines', size=25)
```

```python
df.iplot(x='A', y='B', mode='lines+markers', size=15)
```

```python
cf.datagen.sinwave(10,0.25).iplot(kind='surface')
```

```python
cf.datagen.scatter3d(2,150, mode='stocks').iplot(kind='scatter3d', x='x',y='y', z='z')
```
