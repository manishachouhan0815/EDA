```python
#distplot
sns.distplot(x, kde=True, hist=True, rug=False, bins=30)
#kde= kernel density estimation, if kde=False this curve line will go off 
#hist=histogram, if hist=False we cannot se the hist only curve line if kde=True
```

```python
f,ax=plt.subplots(figsize= (6,6))
cmap=sns.cubehelix_palette(as_cmap= True, dark=0, light=1, reverse=True)
sns.kdeplot(x,y,cmap=cmap, n_levels=60, shade=True)
```

```python
g=sns.jointplot(x,y, kind='kde', color='r')
g.plot_joint(plt.scatter, c='b', s=30, linewidth=1, marker='+')
g.ax_joint.collections[0].set_alpha(0)
```

## 5) Controling PLoted Figure Aesthetics 

```python
def sinplot():
    x=np.linspace(0,14,100)
    for i in range(1,5):
        plt.plot(x,np.sin(x+i*0.5)*(7-i))
```

```python
sinplot()
```

```python
sns.set_style('ticks', {'axes.grid': True, 'xtick.direction': 'in'})
sinplot()
sns.despine(left = True, bottom= False)
```
