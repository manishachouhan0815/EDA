```python
#compute the x and y coordinates for points on sine curve 
np.arange(0,4*np.pi,0.1) # range is from 0 to 4*3.144 with step size 0.1
```

```python
x=np.arange(0, 4*np.pi, 0.1)
y=np.sin(x)
plt.title('sine wave form')
plt.plot(x,y)
plt.show()
```

```python
#Subplot()
#compute the x and y coordinates for points on sin and cosine curves
x=np.arange(0,5*np.pi, 0.1)
y_sin=np.sin(x)
y_cos=np.cos(x)
#set up a subplot grid that has height 2 and width 1,
#and set the first such subplot as active.
plt.subplot(2,1,1)

#make the first plot
plt.plot(x,y_sin,'r--')
plt.title('Sine')

#set the second subplot as active, and make the second plot
plt.subplot(2,1,2)
plt.plot(x,y_cos,'g--')
plt.title('Cosine')

#show the figure
plt.show
```

```python
#PIE chart 
#data to plot 
labels='Python','C++','Ruby','Java'
sizes=[215,130,245,210]
colors=['gold','yellowgreen','lightcoral','lightskyblue']
explode=(0.4,0,0,0) #explode 1st slice
#plot
plt.pie(sizes,explode=explode, labels=labels, colors=colors,autopct='%1.1f%%', shadow=False)
plt.axis('equal')
plt.show()
```
